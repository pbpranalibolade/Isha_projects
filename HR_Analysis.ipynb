```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
df_hr['BusinessTravel']= label_encoder.fit_transform(df_hr['BusinessTravel']) 
df_hr['Department']= label_encoder.fit_transform(df_hr['Department'])
df_hr['EducationField']= label_encoder.fit_transform(df_hr['EducationField'])
df_hr['Gender']= label_encoder.fit_transform(df_hr['Gender'])
df_hr['JobRole']= label_encoder.fit_transform(df_hr['JobRole'])
df_hr['MaritalStatus']= label_encoder.fit_transform(df_hr['MaritalStatus'])
df_hr['Over18']= label_encoder.fit_transform(df_hr['Over18'])
df_hr['OverTime']= label_encoder.fit_transform(df_hr['OverTime'])
df_hr['Attrition']= label_encoder.fit_transform(df_hr['Attrition'])

print(df_hr.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
target = df_hr['Attrition']
features = df_hr.drop('Attrition', axis = 1)
#create the train/test split
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.4, random_state=10)
#Create the model and train
model = RandomForestClassifier()
model.fit(X_train,y_train)
#predict the results for test
test_pred = model.predict(X_test)
#test the accuracy
accuracy_score(y_test, test_pred)
```
